```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#import data
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
#Merge 2 csv files
data = pd.concat([train,test])
data.shape
```

```python
#Remove unnecessary columns
data.drop(columns=['Unnamed: 0', 'id'], inplace=True)
```

```python
#Count number of customers satisfaction
data['satisfaction'].value_counts()
```

```python
#Checking the nature of data set: balanced or imbalanced?
plt.figure(figsize = (8,5))
data.satisfaction.value_counts(normalize = True).plot(kind ='bar', color = ['red', 'green'], alpha = 0.8, rot = 0)
plt.title('Neutral or Dissatisfied vs Satisfied')
plt.show()
```

```python
#The percentage of customers that satisfied and neutral or dissatisfied
print( round((56428 / (56428 + 73452) * 100), 2), "% of customers satisfied with the flight")
print( round((73452 / (56428 + 73452) * 100), 2), "% of customers neutral or dissatisfied with the flight")
```
